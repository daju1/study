в системе центра масс плотность вероятности того что вектор скорости лежит в интервале углов

от $\varphi$ ... $\varphi + d{\varphi}$ составляет

$\frac{p(\varphi ... \varphi + d{\varphi})}{d{\varphi}} = \frac{1}{2\pi}$

в интервале от $0$ до $2\pi$

Пусть в системе центра масс частица имеет скорость

$$v_{3_x} = v_3 \cdot cos(\varphi)$$

$$v_{3_y} = v_3 \cdot sin(\varphi)$$


тогда в системе лабораторной 

$$v_{3_x} = 3 v + v_3 \cdot cos(\varphi) = 3 v + \frac{2v}{3} \cdot cos(\varphi) = \frac{11v}{3} \cdot cos(\varphi)$$

$$v_{3_y} = v_3 \cdot sin(\varphi) = \frac{2v}{3} \cdot sin(\varphi)$$

угол направления в лабораторной системе , отсчитываемый от направления от первоначальной скорости

$$tg \psi = \frac{v_{3_x}}{v_{3_y}} = \frac{3 v + v_3 \cdot cos(\varphi)}{v_3 \cdot sin(\varphi)} = \frac{11 v \cdot cos(\varphi)}{2 v \cdot sin(\varphi)} = \frac{11}{2}ctg(\varphi)$$
